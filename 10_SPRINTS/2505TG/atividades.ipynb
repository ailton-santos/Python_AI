```
# Função para verificação de número
def verificar_numero(valor):
    return 'P' if valor > 0 else 'N'
    '''A parte valor > 0 é uma expressão booleana.'''

# Testando
numero = float(input("Digite um número: "))
print("Resultado:", verificar_numero(numero))
```

```
# Função para calcular imposto
def somaImposto(taxaImposto, custo):
  imposto = custo * (taxaImposto/100)
  custo_final = custo + imposto
  return custo_final

# Pedindo as informações para o usuário
custo_dig = float(input("Digite o custo do produto: R$ "))
taxa_dig = float(input("Digite a taxa de imposto (%): "))

# Mandando as informações para a funão
total = somaImposto(taxa_dig, custo_dig)
# custo_dig = custo e taxa_dig = taxaImposto

print(f"Valor final com imposto: R$ {total:.2f}")
```

```
# Perguntando quantas pessoas serão cadastradas
n = int(input("Quantas pessoas serão cadastradas? "))

# Variáveis para armazenar soma e quantidade de homens e mulheres
soma_homens = 0
qtd_homens = 0
soma_mulheres = 0
qtd_mulheres = 0

# Variável para armazenar a soma total do grupo
soma_grupo = 0

# Laço para repetir o cadastro n vezes
for i in range(n):
    # Pedindo sexo e idade de cada pessoa para o usuário
    sexo = input("Sexo (M/F): ")
    idade = int(input("Idade: "))

    soma_grupo += idade

    # Verificando se é mulher
    if sexo == "F" or sexo == "f":
        soma_mulheres += idade
        qtd_mulheres += 1

    # Verificando se é homem
    elif sexo == "M" or sexo == "m":
        soma_homens += idade
        qtd_homens += 1

# Calculando média geral do grupo
media_grupo = soma_grupo / n

# Calculando média das mulheres
if qtd_mulheres > 0:
    media_mulheres = soma_mulheres / qtd_mulheres
    print(f"Idade média das mulheres: {media_mulheres:.2f}")
else:
    print("Idade média das mulheres: não há mulheres informadas")

# Calculando média dos homens
if qtd_homens > 0:
    media_homens = soma_homens / qtd_homens
    print(f"Idade média dos homens: {media_homens:.2f}")
else:
    print("Idade média dos homens: não há homens informados")

# Mostrando média total do grupo
print(f"Idade média do grupo: {media_grupo:.2f}")
```
