```python
#dog[0ou1,0ou1,0ou1,0,1]
#pig[0ou1,1,0,1,0]

pig=[[0,1,0,1,0],[1,1,0,1,0]]

total_var=3

total_amount=int(2**total_var)
dog=[]



while len(dog)!=total_amount:
  x=1
  for i in range (int(2**(total_var-x))):
    dog.append([0])
  for i in range (int(2**(total_var-x))):
    dog.append([1])
  x+=1
print('oi',dog)

y=0
while x != total_var+1:
  while y != total_amount:
    for i in range (int(2**(total_var-x))):
      dog[y].append(0)
      y+=1
    for i in range (int(2**(total_var-x))):
      dog[y].append(1)
      y+=1
  y=0
  x+=1

print(dog)

```

```python
for dog_sample in dog:
  dog_sample.append(0)
  dog_sample.append(1)

print(dog)
```

```python
training_x=[]
training_y=[]

for pig_sample in pig:
  training_x.append(pig_sample)
  training_y.append(0)

shuffle(dog)
for dog_sample in dog[:5]:
  training_x.append(dog_sample)
  training_y.append(1)

print(training_x)
```

```python
modelo = LinearSVC()
modelo.fit(training_x,training_y)
```

```python
animal_ = [1,1,1,1,1]
modelo.predict([animal_])
```
